# Contrastive Learning for Health Monitoring: SimCLR Analysis of Oura Ring Data

## Project Overview

This project implements a SimCLR (Simple Contrastive Learning of Representations) framework to analyze health monitoring data from Oura Ring devices. The goal is to learn meaningful representations of daily health patterns and identify illness episodes through unsupervised contrastive learning.

### Key Features
- **Unsupervised Learning**: Uses SimCLR to learn representations without labeled data
- **Health Pattern Recognition**: Identifies illness episodes from wearable device data
- **Advanced Visualization**: 2D/3D UMAP and t-SNE visualizations
- **Statistical Analysis**: Comprehensive feature analysis and effect size calculations

### Dataset
- **Source**: Oura Ring health monitoring data (2020-2024)
- **Features**: 28 health metrics including sleep, activity, heart rate, and temperature
- **Duration**: ~4 years of daily measurements
- **Labels**: Known illness dates for validation

### Methodology
1. **Data Preprocessing**: Z-score standardization and feature selection
2. **Contrastive Learning**: SimCLR with Mahalanobis distance similarity
3. **Representation Learning**: 64-dimensional embeddings
4. **Visualization**: UMAP and t-SNE for pattern discovery
5. **Analysis**: Statistical comparison of healthy vs. illness patterns

````python
# Professional visualization functions
def create_professional_plot(embeddings, labels, title="Health Pattern Analysis", 
                            method="UMAP", n_components=2, figsize=(12, 8)):
    """
    Create professional visualizations of learned embeddings.
    
    Args:
        embeddings: Learned embeddings from SimCLR
        labels: Binary labels (0=healthy, 1=sick)
        title: Plot title
        method: Dimensionality reduction method ('UMAP' or 't-SNE')
        n_components: Number of dimensions (2 or 3)
        figsize: Figure size tuple
    """
    # Perform dimensionality reduction
    if method == "UMAP":
        reducer = umap.UMAP(n_components=n_components, random_state=42)
    else:  # t-SNE
        reducer = TSNE(n_components=n_components, random_state=42)
    
    embeddings_reduced = reducer.fit_transform(embeddings)
    
    # Create figure
    if n_components == 3:
        fig = plt.figure(figsize=figsize)
        ax = fig.add_subplot(111, projection='3d')
        
        # Plot healthy points
        healthy_mask = labels == 0
        ax.scatter(embeddings_reduced[healthy_mask, 0], 
                  embeddings_reduced[healthy_mask, 1], 
                  embeddings_reduced[healthy_mask, 2], 
                  c=COLORS['healthy'], label='Healthy', alpha=0.6, s=30)
        
        # Plot sick points
        sick_mask = labels == 1
        ax.scatter(embeddings_reduced[sick_mask, 0], 
                  embeddings_reduced[sick_mask, 1], 
                  embeddings_reduced[sick_mask, 2], 
                  c=COLORS['sick'], label='Illness Episodes', alpha=0.8, s=60)
        
        ax.set_xlabel('Dimension 1')
        ax.set_ylabel('Dimension 2')
        ax.set_zlabel('Dimension 3')
        
    else:  # 2D
        fig, ax = plt.subplots(figsize=figsize)
        
        # Plot healthy points
        healthy_mask = labels == 0
        ax.scatter(embeddings_reduced[healthy_mask, 0], 
                  embeddings_reduced[healthy_mask, 1], 
                  c=COLORS['healthy'], label='Healthy', alpha=0.6, s=30)
        
        # Plot sick points
        sick_mask = labels == 1
        ax.scatter(embeddings_reduced[sick_mask, 0], 
                  embeddings_reduced[sick_mask, 1], 
                  c=COLORS['sick'], label='Illness Episodes', alpha=0.8, s=60)
        
        ax.set_xlabel('Dimension 1')
        ax.set_ylabel('Dimension 2')
    
    # Styling
    ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
    ax.legend(loc='best', frameon=True, fancybox=True, shadow=True)
    ax.grid(True, alpha=0.3)
    
    # Add statistics
    n_healthy = np.sum(healthy_mask)
    n_sick = np.sum(sick_mask)
    ax.text(0.02, 0.98, f'Healthy: {n_healthy}\nIllness: {n_sick}', 
            transform=ax.transAxes, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    return fig, ax

def plot_training_progress(losses, title="SimCLR Training Progress"):
    """Plot training loss with professional styling."""
    fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.plot(losses, color=COLORS['primary'], linewidth=2, alpha=0.8)
    ax.fill_between(range(len(losses)), losses, alpha=0.3, color=COLORS['primary'])
    
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Contrastive Loss')
    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    # Add final loss annotation
    final_loss = losses[-1]
    ax.annotate(f'Final Loss: {final_loss:.4f}', 
                xy=(len(losses)-1, final_loss), 
                xytext=(len(losses)*0.7, max(losses)*0.8),
                arrowprops=dict(arrowstyle='->', color='red', alpha=0.7),
                fontsize=12, ha='center',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))
    
    plt.tight_layout()
    return fig, ax

print("ðŸŽ¨ Professional visualization functions loaded!")
````

````python
# Error handling and validation functions
def validate_data(data, required_columns=None):
    """
    Validate input data for common issues.
    
    Args:
        data: DataFrame to validate
        required_columns: List of required column names
    
    Returns:
        dict: Validation results
    """
    validation_results = {
        'is_valid': True,
        'warnings': [],
        'errors': []
    }
    
    # Check if data is empty
    if data.empty:
        validation_results['errors'].append("Data is empty")
        validation_results['is_valid'] = False
        return validation_results
    
    # Check for required columns
    if required_columns:
        missing_cols = set(required_columns) - set(data.columns)
        if missing_cols:
            validation_results['errors'].append(f"Missing columns: {missing_cols}")
            validation_results['is_valid'] = False
    
    # Check for excessive missing values
    missing_pct = data.isnull().sum() / len(data) * 100
    high_missing = missing_pct[missing_pct > 50].index.tolist()
    if high_missing:
        validation_results['warnings'].append(f"High missing values (>50%) in: {high_missing}")
    
    # Check for constant columns
    constant_cols = data.columns[data.nunique() <= 1].tolist()
    if constant_cols:
        validation_results['warnings'].append(f"Constant columns detected: {constant_cols}")
    
    # Check for infinite values
    inf_cols = data.columns[np.isinf(data.select_dtypes(include=[np.number])).any()].tolist()
    if inf_cols:
        validation_results['warnings'].append(f"Infinite values in: {inf_cols}")
    
    return validation_results

def safe_train_simclr(features, epochs=100, batch_size=64, lr=1e-3, max_retries=3):
    """
    Train SimCLR with error handling and automatic retries.
    
    Args:
        features: Input feature matrix
        epochs: Number of training epochs
        batch_size: Batch size for training
        lr: Learning rate
        max_retries: Maximum number of retry attempts
    
    Returns:
        tuple: (encoder, losses, success_flag)
    """
    print(f"ðŸš€ Starting SimCLR training with {epochs} epochs...")
    
    for attempt in range(max_retries):
        try:
            print(f"ðŸ“Š Training attempt {attempt + 1}/{max_retries}")
            
            # Validate input
            if features is None or len(features) == 0:
                raise ValueError("Empty feature matrix")
            
            if np.isnan(features).any():
                print("âš ï¸ NaN values detected, replacing with zeros...")
                features = np.nan_to_num(features, nan=0.0)
            
            if np.isinf(features).any():
                print("âš ï¸ Infinite values detected, replacing with zeros...")
                features = np.nan_to_num(features, posinf=0.0, neginf=0.0)
            
            # Train model
            encoder, losses = train_simclr_clean(features, epochs, batch_size, lr)
            
            # Validate results
            if len(losses) == 0:
                raise ValueError("No losses recorded")
            
            if np.isnan(losses).any():
                raise ValueError("NaN losses detected")
            
            print("âœ… Training completed successfully!")
            return encoder, losses, True
            
        except Exception as e:
            print(f"âŒ Training attempt {attempt + 1} failed: {e}")
            if attempt == max_retries - 1:
                print("ðŸ’¥ All training attempts failed!")
                return None, None, False
            else:
                print("ðŸ”„ Retrying with adjusted parameters...")
                # Adjust parameters for retry
                batch_size = max(16, batch_size // 2)
                lr = lr * 0.5

def validate_embeddings(embeddings):
    """
    Validate learned embeddings for quality and consistency.
    
    Args:
        embeddings: Learned embedding matrix
    
    Returns:
        dict: Validation results
    """
    results = {
        'is_valid': True,
        'quality_score': 0.0,
        'warnings': [],
        'errors': []
    }
    
    # Check for NaN values
    if np.isnan(embeddings).any():
        results['errors'].append("NaN values in embeddings")
        results['is_valid'] = False
    
    # Check for infinite values
    if np.isinf(embeddings).any():
        results['errors'].append("Infinite values in embeddings")
        results['is_valid'] = False
    
    # Check embedding quality (variance)
    embedding_var = np.var(embeddings, axis=0)
    low_var_dims = np.sum(embedding_var < 1e-6)
    if low_var_dims > embeddings.shape[1] * 0.5:
        results['warnings'].append(f"Low variance in {low_var_dims} dimensions")
    
    # Calculate quality score
    if results['is_valid']:
        # Higher variance = better quality
        avg_var = np.mean(embedding_var)
        results['quality_score'] = min(1.0, avg_var / 10.0)  # Normalize to [0,1]
    
    return results

print("ðŸ›¡ï¸ Error handling and validation functions loaded!")
````

````python
# Comprehensive results analysis functions
def analyze_feature_importance(healthy_data, sick_data, feature_names, top_k=10):
    """
    Analyze feature importance using effect sizes (Cohen's d).
    
    Args:
        healthy_data: DataFrame of healthy samples
        sick_data: DataFrame of sick samples
        feature_names: List of feature names
        top_k: Number of top features to return
    
    Returns:
        dict: Analysis results
    """
    effect_sizes = {}
    
    for feature in feature_names:
        if feature in healthy_data.columns and feature in sick_data.columns:
            healthy_mean = healthy_data[feature].mean()
            sick_mean = sick_data[feature].mean()
            
            # Pooled standard deviation
            healthy_std = healthy_data[feature].std()
            sick_std = sick_data[feature].std()
            pooled_std = np.sqrt((healthy_std**2 + sick_std**2) / 2)
            
            # Cohen's d effect size
            effect_size = abs(healthy_mean - sick_mean) / pooled_std
            effect_sizes[feature] = effect_size
    
    # Sort by effect size
    sorted_effects = sorted(effect_sizes.items(), key=lambda x: x[1], reverse=True)
    
    return {
        'top_features': sorted_effects[:top_k],
        'all_effects': effect_sizes,
        'summary': {
            'total_features': len(feature_names),
            'significant_features': len([e for e in effect_sizes.values() if e > 0.5]),
            'large_effects': len([e for e in effect_sizes.values() if e > 0.8])
        }
    }

def analyze_clustering_quality(embeddings, labels, sick_dates, date_column):
    """
    Analyze the quality of learned representations for illness clustering.
    
    Args:
        embeddings: Learned embeddings
        labels: Binary labels (0=healthy, 1=sick)
        sick_dates: List of known sick dates
        date_column: Date column from original data
    
    Returns:
        dict: Clustering quality analysis
    """
    from sklearn.metrics import silhouette_score
    from sklearn.cluster import KMeans
    
    # Calculate silhouette score
    try:
        silhouette = silhouette_score(embeddings, labels)
    except:
        silhouette = 0.0
    
    # Analyze sick day clustering
    sick_indices = np.where(labels == 1)[0]
    healthy_indices = np.where(labels == 0)[0]
    
    if len(sick_indices) > 0:
        # Calculate centroid distances
        sick_centroid = np.mean(embeddings[sick_indices], axis=0)
        healthy_centroid = np.mean(embeddings[healthy_indices], axis=0)
        
        # Distance from each sick point to sick centroid
        sick_distances = [np.linalg.norm(emb - sick_centroid) for emb in embeddings[sick_indices]]
        healthy_distances = [np.linalg.norm(emb - healthy_centroid) for emb in embeddings[healthy_indices]]
        
        # Clustering quality metrics
        sick_cohesion = np.mean(sick_distances)
        healthy_cohesion = np.mean(healthy_distances)
        separation = np.linalg.norm(sick_centroid - healthy_centroid)
        
        # Identify well vs poorly clustered sick days
        sick_dates_series = pd.to_datetime(date_column)
        well_clustered = []
        poorly_clustered = []
        
        for i, sick_idx in enumerate(sick_indices):
            date = sick_dates_series.iloc[sick_idx]
            distance = sick_distances[i]
            
            # Threshold based on median distance
            if distance < np.median(sick_distances):
                well_clustered.append(date.strftime('%Y-%m-%d'))
            else:
                poorly_clustered.append(date.strftime('%Y-%m-%d'))
        
        return {
            'silhouette_score': silhouette,
            'sick_cohesion': sick_cohesion,
            'healthy_cohesion': healthy_cohesion,
            'separation': separation,
            'well_clustered_dates': well_clustered,
            'poorly_clustered_dates': poorly_clustered,
            'clustering_quality': 'Good' if silhouette > 0.3 else 'Fair' if silhouette > 0.1 else 'Poor'
        }
    
    return {'error': 'No sick days found'}

def generate_analysis_report(embeddings, labels, data, feature_names, sick_dates):
    """
    Generate a comprehensive analysis report.
    
    Args:
        embeddings: Learned embeddings
        labels: Binary labels
        data: Original dataset
        feature_names: List of feature names
        sick_dates: List of sick dates
    
    Returns:
        dict: Comprehensive analysis report
    """
    print("ðŸ“Š Generating comprehensive analysis report...")
    
    # Separate healthy and sick data
    healthy_mask = labels == 0
    sick_mask = labels == 1
    
    healthy_data = data[healthy_mask][feature_names]
    sick_data = data[sick_mask][feature_names]
    
    # Feature importance analysis
    feature_analysis = analyze_feature_importance(healthy_data, sick_data, feature_names)
    
    # Clustering quality analysis
    clustering_analysis = analyze_clustering_quality(embeddings, labels, sick_dates, data['date'])
    
    # Embedding validation
    embedding_validation = validate_embeddings(embeddings)
    
    # Summary statistics
    summary = {
        'total_samples': len(data),
        'healthy_samples': np.sum(healthy_mask),
        'sick_samples': np.sum(sick_mask),
        'embedding_dimensions': embeddings.shape[1],
        'feature_count': len(feature_names),
        'sick_dates_count': len(sick_dates)
    }
    
    # Key findings
    findings = {
        'top_discriminative_features': [f[0] for f in feature_analysis['top_features'][:5]],
        'clustering_quality': clustering_analysis.get('clustering_quality', 'Unknown'),
        'embedding_quality_score': embedding_validation['quality_score'],
        'silhouette_score': clustering_analysis.get('silhouette_score', 0.0)
    }
    
    report = {
        'summary': summary,
        'findings': findings,
        'feature_analysis': feature_analysis,
        'clustering_analysis': clustering_analysis,
        'embedding_validation': embedding_validation,
        'recommendations': generate_recommendations(findings, clustering_analysis)
    }
    
    return report

def generate_recommendations(findings, clustering_analysis):
    """Generate recommendations based on analysis results."""
    recommendations = []
    
    if findings['silhouette_score'] < 0.1:
        recommendations.append("Consider adjusting SimCLR parameters (temperature, learning rate)")
        recommendations.append("Try different data augmentation strategies")
    
    if findings['embedding_quality_score'] < 0.3:
        recommendations.append("Increase training epochs or adjust model architecture")
        recommendations.append("Check for data quality issues")
    
    if len(findings['top_discriminative_features']) < 3:
        recommendations.append("Feature selection may be too restrictive")
        recommendations.append("Consider including more health metrics")
    
    if clustering_analysis.get('clustering_quality') == 'Poor':
        recommendations.append("Illness episodes may be too heterogeneous")
        recommendations.append("Consider temporal modeling for disease progression")
    
    return recommendations

print("ðŸ“ˆ Comprehensive analysis functions loaded!")
````

### SimCLR Architecture

The implementation uses a **Simple Contrastive Learning of Representations (SimCLR)** framework specifically adapted for health monitoring data:

#### 1. **Data Augmentation**
- **Gaussian Noise**: `x + N(0, 0.1Â²)` for robustness
- **Feature Dropout**: Random masking with p=0.1 for regularization
- **Contrastive Pairs**: Generate positive pairs for training

#### 2. **Encoder Architecture**

- **Hidden Layer**: 256 neurons with ReLU activation
- **Output Dimension**: 64-dimensional embeddings

- **InfoNCE Loss**: Contrastive learning objective
- **Temperature Scaling**: Ï„=0.5 for similarity calibration

- **Optimizer**: Adam with lr=1e-3
- **Batch Size**: 64 (with automatic adjustment)
- **Epochs**: 100 (with early stopping)
- **Device**: CUDA if available, CPU fallback

## Mahanalobis distance

````python
# 5. SimCLR Loss Function with Mahalanobis Distance
def mahalanobis_similarity(z1, z2, inv_cov):
    diff = z1.unsqueeze(1) - z2.unsqueeze(0)  # shape: (2N, 2N, D)
    sim = -torch.einsum('bij,jk,bik->bi', diff, inv_cov, diff)  # negative squared Mahalanobis distance
    return sim

def simclr_loss(z1, z2, temperature=0.5):
    z = torch.cat([z1, z2], dim=0)  # shape: (2N, D)
    z = (z - z.mean(dim=0)) / (z.std(dim=0) + 1e-8)  # normalize with larger epsilon

    # Estimate covariance and add stronger regularization
    B, D = z.shape
    cov = (z.T @ z) / (B - 1)  # (D, D)
    
    # Add stronger regularization to make matrix well-conditioned
    reg = 1e-2 * torch.eye(D, device=z.device)  # Increased from 1e-3 to 1e-2
    regularized_cov = cov + reg
    
    # Use try-catch to fallback to identity if still fails
    try:
        inv_cov = torch.linalg.pinv(regularized_cov)
    except:
        # Fallback to identity matrix (equivalent to Euclidean distance)
        inv_cov = torch.eye(D, device=z.device)
    
    # Compute Mahalanobis similarities
    def mahalanobis_sim(a, b):
        diff = a.unsqueeze(1) - b.unsqueeze(0)  # (B, B, D)
        sim = -torch.einsum('bij,jk,bik->bi', diff, inv_cov, diff)  # negative squared Mahalanobis
        return sim

    sim_matrix = mahalanobis_sim(z, z)
    N = z1.size(0)

    mask = torch.eye(2 * N, device=z.device).bool()
    sim_matrix.masked_fill_(mask, float('-inf'))

    # Positive pairs
    pos_sim = -torch.einsum('bi,ij,bj->b', z1 - z2, inv_cov, z1 - z2)
    pos_sim = torch.cat([pos_sim, pos_sim], dim=0)

    numerator = torch.exp(pos_sim / temperature)
    denominator = torch.sum(torch.exp(sim_matrix / temperature), dim=1)
    loss = -torch.log(numerator / denominator)
    return loss.mean()

# SimCLR-style Contrastive Learning for Oura Data

# 1. Imports
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D


# 2. Dataset Preparation
# (Assume zscore_df is already loaded with the scaled columns and no missing values)
cols_to_use = ['Temperature Trend Deviation', 'Total Burn', 'Long Periods of Inactivity', 'Inactive Time',
               'Non-wear Time', 'Temperature Deviation (Â°C)', 'Steps', 'Deep Sleep Duration',
               'Low Activity Time', 'Average HRV', 'Respiratory Rate', 'Activity Burn',
               'Total Sleep Duration', 'Sleep Latency', 'Equivalent Walking Distance', 'Sleep Efficiency',
               'Medium Activity Time', 'Lowest Resting Heart Rate', 'Average Resting Heart Rate', 'Awake Time',
               'Sleep Timing', 'Light Sleep Duration', 'Rest Time', 'High Activity Time', 'REM Sleep Duration',
               'Average MET', 'Total Bedtime ', 'Restless Sleep']

zscore_df = zscore_df.dropna(subset=cols_to_use)
features = zscore_df[cols_to_use].values.astype(np.float32)

# 3. Augmentation Functions
def add_noise(x):
    return x + np.random.normal(0, 0.1, size=x.shape).astype(np.float32)

def feature_dropout(x):
    mask = np.random.binomial(1, 0.1, size=x.shape).astype(bool)
    x[mask] = 0
    return x.astype(np.float32)

def contrastive_augment(x):
    return add_noise(x.copy()), feature_dropout(x.copy())

# 4. Encoder Function
import torch.nn as nn

def build_encoder(input_dim, output_dim=64):
    return nn.Sequential(
        nn.Linear(input_dim, 256),
        nn.ReLU(),
        nn.Linear(256, output_dim)
    )

# 6. Training Loop
def train_simclr(features, epochs=100, batch_size=64, lr=1e-3):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    encoder = build_encoder(features.shape[1]).to(device)
    optimizer = torch.optim.Adam(encoder.parameters(), lr=lr)
    losses = []

    for epoch in range(epochs):
        encoder.train()
        total_loss = 0
        indices = np.random.permutation(len(features))
        for i in range(0, len(features), batch_size):
            batch_idx = indices[i:i+batch_size]
            batch = features[batch_idx]
            x1_list, x2_list = zip(*[contrastive_augment(x) for x in batch])
            x1 = torch.tensor(np.stack(x1_list), dtype=torch.float32).to(device)
            x2 = torch.tensor(np.stack(x2_list), dtype=torch.float32).to(device)

            z1 = encoder(x1)
            z2 = encoder(x2)
            loss = simclr_loss(z1, z2)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        avg_loss = total_loss / (len(features) / batch_size)
        losses.append(avg_loss)
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")

    return encoder, losses

# 7. Run Training
encoder_model, training_losses = train_simclr(features)

# train_simclr_debug(features)

# 8. Plot Training Loss
plt.plot(training_losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('SimCLR Training Loss')
plt.show()

# 9. Extract Embeddings
encoder_model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
x_tensor = torch.tensor(features, dtype=torch.float32).to(device)
with torch.no_grad():
    embeddings = encoder_model(x_tensor).cpu().numpy()

# 10. Create Sick Labels
# sick_dates = ['2020-09-22','2021-01-10', '2021-01-11','2021-01-12', '2021-01-13', '2021-01-15', '2021-01-16', '2021-01-17',  
#     '2021-04-03', '2021-04-10', '2022-07-10',  '2022-07-16', '2022-07-17', '2022-07-18', '2022-07-19', 
#     '2022-11-02', '2022-11-05', '2024-01-12', '2023-12-11',  '2024-07-11', '2022-12-22', '2021-12-07', '2021-11-19', '2021-12-12', '2021-12-27', '2023-10-05']
sick_dates = ['2021-01-10', '2021-01-11','2021-01-12', '2021-01-13', '2021-01-15', '2021-01-16', '2021-01-17',  
    '2021-04-03', '2021-04-10', '2022-07-10',  '2022-07-16', '2022-07-17', '2022-07-18', '2022-07-19', 
    '2022-11-02', '2022-11-05', '2024-01-12']


zscore_df['sick'] = zscore_df['date'].astype(str).isin(sick_dates).astype(int)
labels = zscore_df['sick'].values

# # 11. 3D UMAP Visualization
# umap_3d = umap.UMAP(n_components=3, random_state=42)
# emb_umap = umap_3d.fit_transform(embeddings)

# fig = plt.figure(figsize=(10, 7))
# ax = fig.add_subplot(111, projection='3d')
# ax.scatter(emb_umap[labels==0, 0], emb_umap[labels==0, 1], emb_umap[labels==0, 2], label='Healthy', alpha=0.5)
# ax.scatter(emb_umap[labels==1, 0], emb_umap[labels==1, 1], emb_umap[labels==1, 2], label='Sick', alpha=0.8, color='red')
# ax.set_title('3D UMAP of Learned Embeddings')
# ax.set_xlabel('Dim 1')
# ax.set_ylabel('Dim 2')
# ax.set_zlabel('Dim 3')
# ax.legend()
# plt.show()

# 11. 3D UMAP Visualization with Date Labels for Sick Days
umap_3d = umap.UMAP(n_components=3, random_state=42)
emb_umap = umap_3d.fit_transform(embeddings)

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot healthy points
ax.scatter(emb_umap[labels==0, 0], emb_umap[labels==0, 1], emb_umap[labels==0, 2], 
           label='Healthy', alpha=0.5, s=20)

# Plot sick points
sick_indices = np.where(labels==1)[0]
ax.scatter(emb_umap[labels==1, 0], emb_umap[labels==1, 1], emb_umap[labels==1, 2], 
           label='Sick', alpha=0.8, color='red', s=50)

# Add date labels for sick points only
sick_dates_formatted = zscore_df[zscore_df['sick']==1]['date'].dt.strftime('%Y-%m-%d').values

for i, date in enumerate(sick_dates_formatted):
    sick_idx = sick_indices[i]
    ax.text(emb_umap[sick_idx, 0], emb_umap[sick_idx, 1], emb_umap[sick_idx, 2], 
            date, fontsize=8, alpha=0.8)

ax.set_title('3D UMAP of Learned Embeddings with Sick Day Labels')
ax.set_xlabel('Dim 1')
ax.set_ylabel('Dim 2')
ax.set_zlabel('Dim 3')
ax.legend()
plt.tight_layout()
plt.show()
````
